```python
import cv2 as a
import numpy as n
from matplotlib import pyplot as p
```

```python
img=a.imread("cat.jpeg",0)
p.imshow(img)
```

```python
canny=a.Canny(img,80,200)
p.imshow(canny)
```

# frame capture

```python
ca=a.VideoCapture(0)
```

```python
while(1):
    ret,frame=ca.read()
    gray=a.cvtColor(frame,a.COLOR_BGR2GRAY)
    a.imshow('original',frame)
    edges=a.Canny(frame,100,200)
    a.imshow('edges',edges)
    k=a.waitKey(5)&0xFF
    if k==27:
        break
ca.release()
a.destroyAllWindows()
```

# captered video

```python
cap=a.VideoCapture("demon.mp4")
```

```python
while(cap.isOpened()):
    ret,frames=cap.read()
    gray=a.cvtColor(frames,a.COLOR_BGR2GRAY)
    a.imshow('original',frames)
    edge=a.Canny(frames,100,200)
    a.imshow("edge",edge)
    k=a.waitKey(5)&0xFF
    if k==27:
        break
cap.release()
a.destroyAllWindows()
```
